```python
import pandas as pd

df = pd.read_csv('../data/annotations_2017_A_fixed.csv')
```

```python
df.isna().sum()
```

```python
record_lengths = pd.Series()
c = 0
for col in df.columns:
    j = 0
    for i in df[col]:
        j += 1
        if i != i:
            record_lengths[c] = j
            break
    c += 1
record_lengths
```

```python
record_lengths.mean() / 60
```

```python
record_lengths.max() / 60
```

```python
record_lengths.min() / 60
```

```python
record_lengths.sum() / 60
```

```python
len(df)
```

```python
segments_all = []
for col in df.columns:
    segments_in_col = []
    segment = []
    for i in range(len(df)):
        if df[col][i] == 1:
            if len(segment) == 0:  # Only append start point if it hasn't been added yet
                segment.append(i)
        elif df[col][i] != 1 and len(segment) > 0:  # Only append end point if a start point has been added
            segment.append(i)
            segments_in_col.append(segment)
            segment = []
    segments_all.append(segments_in_col)
```

```python
segments_all
```

```python
len(segments_all)
```

```python
lengths = [len(x) for x in segments_all]
lengths
```

```python
sum(lengths)
```

```python
# iterate through segments_all and get the difference between the start and end points for each segment
durations = []
for col in segments_all:
    for segment in col:
        durations.append(segment[1] - segment[0])
durations
```

```python
sum(durations) / 60
```

```python
record_lengths.sum() / 60
```

```python
print( f'{( ( sum(durations) / 60 ) / ( record_lengths.sum() / 60 ) * 100):.2f} % of recordings are seizures'  )
```
